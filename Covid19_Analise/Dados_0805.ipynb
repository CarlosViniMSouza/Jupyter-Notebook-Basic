```python
#Bibliotecas para anallisar e treinar dados :
import pandas as pd # Visualização de dados
import numpy as np # Cálculos vetorias e matriciais
import seaborn as sns # Biblioteca do matplotlib especifíco 
import matplotlib.pyplot as mtp # Biblioteca para plotagem e gráficos
```

```python
covid_data = pd.read_csv('WHO-COVID-19-global-08-05.csv')
covid_data
```

```python
covid_data.columns = ["Dia", "UF", "País", "Região", "Mortes", "Mortes_Acumula", "Confirmados", "Confirmados_Acumula"]
covid_data
```

```python
#vamos reduzir os dados exclusivamente para o Brasil :
data_covid_brasil = covid_data[covid_data['País']=='Brazil']
data_covid_brasil
```

```python
data_covid_brasil.describe()
```

```python
# Para melhorar a análise, reduziremos o tempo de estudo dos dados: 
data_covid_semana_br = data_covid_brasil.tail(7)
data_covid_semana_br
```

```python
data_covid_semana_br.describe()
```

```python
sns.set(style='ticks')
data_covid_semana_br['Confirmados'].plot(kind='hist')
```

## Vejamos outros tipos de gráficos para demonstrar melhor a situação estudada:

```python
sns.relplot(x="Dia", y="Confirmados", data=data_covid_semana_br)
```

```python
sns.relplot(x="Dia", y="Confirmados", estimator=None, kind="line", data=data_covid_semana_br);
```

```python
sns.set(style="whitegrid")

# Inicializando figura de matplotlib
f, ax = mtp.subplots(figsize=(30,20))

# Plotagem de casos confirmados:
sns.set_color_codes("colorblind")
sns.barplot(x="Dia", y="Confirmados", data=data_covid_semana_br,
           label="Casos Confirmados", color="b")

# Plotagem de obitos registrados:
sns.set_color_codes("deep")
sns.barplot(x="Dia", y="Mortes", data=data_covid_semana_br,
           label="Obitos", color="b")

# Add uma legenda e informações:
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(3,18), ylabel="Mortes", xlabel="Dia")
sns.despine(left=True, bottom=True)
```
